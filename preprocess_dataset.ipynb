```python
from pathlib import Path
import pandas as pd
import math
```

```python
INPUT_TEXT = "./docs/lyrics.csv"
OUTPUT_FOLDER = "./docs/lyrics"
```

```python
Path(OUTPUT_FOLDER).mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(INPUT_TEXT)
df
```

```python
filtered_df = df.iloc[df['lyrics'].dropna().index]
filtered_df = filtered_df.drop(columns=["Unnamed: 0"])
filtered_df
```

```python
for idx, row in filtered_df.iterrows():
    filename = f"{OUTPUT_FOLDER}/{idx}.txt"
    with open(filename, "w") as f:
        f.write(row['lyrics'])
        f.write('\n')
```
